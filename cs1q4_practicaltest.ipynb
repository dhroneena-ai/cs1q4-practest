```python
def analyze_scores():
    name = input("Student Name: ")
    num = int(input("Number of quizzes: "))

    scores = [float(input(f"Score {i+1}:")) for i in range(num)]

    total = sum(scores)
    avg = total / num
    result = "Passed" if avg >= 60 else "Failed"

    print(f"name: {name}, total: {total}, average: {avg}%, {result}")

analyze_scores()
```
